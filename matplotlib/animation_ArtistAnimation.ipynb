```python
"""
ãŸã‘ã®ã“æˆé•·è¨˜éŒ²
"""
# Jupyterä¸Šã§ã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³ãŒå†ç”Ÿå¯èƒ½ã«ã™ã‚‹ãŸã‚ã€matplotlibã®nbaggã‚’æœ‰åŠ¹
%matplotlib nbagg

import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure() #ã€€è¡¨ä½œæˆ
plt.xlim(0, 6) #ã€€xè»¸ã®ç¯„å›²
plt.ylim(0, 10) #ã€€yè»¸ã®ç¯„å›²

ims = [] # ã‚°ãƒ©ãƒ•æç”»é…åˆ—
arr = [0, 0, 0, 0, 0, 0, 0] # ã‚°ãƒ©ãƒ•é…åˆ—åˆæœŸå€¤

for i in range(11): # i = 0ï½ž10ã¾ã§ãƒ«ãƒ¼ãƒ—
    arr[3] = i # æˆé•·å…·åˆ
    im = plt.plot(arr) # æˆé•·è¨˜éŒ²ã®ãƒ—ãƒ­ãƒƒãƒˆå–å¾—
    ims.append(im) # ã‚°ãƒ©ãƒ•æç”»é…åˆ—ã«è¿½åŠ 

# 11æžšã®ãƒ—ãƒ­ãƒƒãƒˆ(æˆé•·è¨˜éŒ²)ã‚’ 300ms ã”ã¨ã«ã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³è¡¨ç¤º
ani = animation.ArtistAnimation(fig, ims, interval=300)

ani.save("img/animation_ArtistAnimation.gif", writer="imagemagick") #ã‚«ãƒ¬ãƒ³ãƒˆãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªã«GIFãƒ•ã‚¡ã‚¤ãƒ«å‡ºåŠ›
plt.show()
```
